```python
import pandas as pd
import numpy as np

# Data Split
# Testing dan Training data
from sklearn.model_selection import train_test_split 

# Data Transforming
# Standarisasi
from sklearn.preprocessing import StandardScaler 

# Normalisasi
from sklearn.preprocessing import MinMaxScaler

# Data Cleaning
# Subtitusi nilai
from sklearn.impute import SimpleImputer 

# Data Encoding
# Onehot
from sklearn.preprocessing import OneHotEncoder
```

File CSV

```python
dataset_pes = pd.read_csv('deets-updated.csv')
dataset_pes.head()
```

```python
dataset_pes.info()
```

#### DATA SPLIT
Metode evaluasi model ini membagi dataset menjadi dua bagian untuk training data dan untuk testing data dengan proporsi tertentu.

##### Bagi dataset menjadi training set dan testing set dengan proporsi 70:30

```python
data = dataset_pes[['height','weight','age','rating_stars']]
label = dataset_pes['team_name']

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=.3)
```

```python
print(f'panjang data training {len(data_train)}',
     f'panjang label training {len(label_train)}',
     f'panjang data testing {len(data_test)}',
     f'panjang label testing {len(label_test)}',
     sep='\n')
```

#### DATA TRANSFORMASI
merubah skala data kedalam bentuk lain. Teknik penskalaan yang biasa dipakai yaitu normalisasi dan standardisasi

##### Lakukan normalisasi data pada salah satu attribute menggunakan Min Max scaler (buatlah copy dataset terlebih dahulu)

```python
#Normalisasi
dataset_pes_copy = dataset_pes.copy()

scaler = MinMaxScaler()

normalize_dataset = scaler.fit_transform(dataset_pes_copy[['height', 'weight', 'age']])

normalize_dataset = pd.DataFrame(normalize_dataset)
normalize_dataset.rename(columns={ 0 :'height', 1 : 'weight', 2 : 'age'}, inplace=True)

normalize_dataset.head(5)
```

##### Lakukan standarisasi pada dataset (buatlah copy dataset terlebih dahulu)

```python
#Standarisasi
dataset_pes_copy.drop('team_name', axis=1, inplace=True)

standart_std = StandardScaler()

scaled_data = standart_std.fit_transform(dataset_pes_copy[['height', 'weight', 'age']])
scaled_data = pd.DataFrame(scaled_data)
scaled_data.rename(columns={ 0 :'height', 1 : 'weight', 2 : 'age'}, inplace=True)

scaled_data_pes = standart_std.fit_transform(dataset_pes[['height', 'weight', 'age']])
scaled_data_pes = pd.DataFrame(scaled_data_pes)
scaled_data_pes.rename(columns={ 0 :'height', 1 : 'weight', 2 : 'age'}, inplace=True)

print('Dataset sebelum scalling:',  # SEBELUM
      dataset_pes.drop('team_name', axis=1).head(6),'\n'
      f'Nilai standar deviasi:\n{np.std(scaled_data_pes)}',
      sep='\n', end='\n\n')

print('Dataset setelah scalling:',  # SESUDAH
      scaled_data[:6],
      f'Nilai standar deviasi: {np.std(scaled_data)}',
      sep='\n')
```

#### DATA CLEANING
Data cleaning adalah memperbarui, membetulkan, dan membuat data menjadi lebih rapi.

##### Lakukan Data cleaning pada data dengan nilai null .Ganti nilai null sesuai ketentuan.( bilangan desimal : mean)

```python
# mengecek nilai null
nulls = dataset_pes_copy.isnull().sum().to_frame()
for index, row in nulls.iterrows():
    print(index +" = ", row[0])
```

```python
# bikin alat
imputer_mean = SimpleImputer(strategy = 'mean')
```

```python
#info
dataset_pes_copy.info(verbose=True, show_counts=True)
```

```python
#mean
dataset_pes_copy['LWF'] = imputer_mean.fit_transform(dataset_pes_copy[['LWF']])
dataset_pes_copy['SS'] = imputer_mean.fit_transform(dataset_pes_copy[['SS']])
dataset_pes_copy['CF'] = imputer_mean.fit_transform(dataset_pes_copy[['CF']])
dataset_pes_copy['RWF'] = imputer_mean.fit_transform(dataset_pes_copy[['RWF']])
```

```python
#nilai null
nulls = dataset_pes_copy.isnull().sum().to_frame()
for index, row in nulls.iterrows():
    print(index +" = ", row[0])
```

#### ENCODING
Encoding adalah teknik yang mengubah variabel kategorikal menjadi variabel numerik. Teknik encoding yang paling terkenal adalah Ordinal Encoding dan One-Hot Encoding

##### Lakukan one hot encoding pada dataset yang kalian miliki.

```python
#copydataset
dataset_pes_copy2 = dataset_pes.copy()
```

```python
# buat alat
encoder_oneHot = OneHotEncoder(sparse=False)

temp_dataset = encoder_oneHot.fit_transform(dataset_pes_copy2[['region']]).astype('int32')
```

```python
# format dataset ke dataframe
temp_dataset = pd.DataFrame(temp_dataset)
```

```python
# hapus label sebelumnya attribute
dataset_pes_copy2.drop('region', axis=1, inplace=True)
```

```python
# tambahkan label baru attribute
dataset_pes_copy2 = dataset_pes_copy2.join(temp_dataset)
```

```python
# ganti nama label attribute 
dataset_pes_copy2.rename(columns = {0:'Africa', 1:'Asia-Oceania', 2:'Europe', 3 : 'North & Central America', 4 : 'South America'}, inplace=True)
```
