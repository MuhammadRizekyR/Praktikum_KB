```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

FILE CSV

```python
pd.set_option("display.max_columns", None)
dataset_pes = pd.read_csv('deets-updated.csv')
dataset_pes.head(20)
```

Attribute Di Dataset

```python
dataset_pes.keys().tolist()
```

Grafik Batang untuk memvisualisasi data kategorikal

```python
plt.figure(figsize = (4,6))
dataset_pes['rating_stars'].value_counts().plot.bar()
plt.title('Grafik Batang',fontsize = 25 , fontweight = 'bold')
plt.xlabel('Rating Stars')
plt.ylabel('Sum Rating Stars')
plt.show()
```

Pemain dengan rating tertinggi lebih sedikit dibanding dengan pemain yang mendapatkan rating terendah

Grafik Pie cocok untuk melihat distribusi

```python
plt.figure(figsize=(4,6))
plt.title('Grafik Pie',fontsize = 25 , fontweight = 'bold')
plt.pie(dataset_pes['foot'].value_counts(), labels=['Right Foot', 'Left Foot'], startangle=90, autopct='%.1f%%', explode=(0.1,0))
plt.legend(loc='lower right')
plt.show()
```

Presentase pemain dengan kaki kanan terkuat lebih besar 57,8% dibanding dengan pemain dengan kaki kiri

Grafik Scatter Plot menunjukan hubungan antara dua variabel

```python
plt.scatter(dataset_pes['registered_position'], dataset_pes['height'], label = 'Tinggi',color  = 'red')
plt.title('Grafik Scatter Plot',fontsize = 25 , fontweight = 'bold')
plt.xticks(rotation = 90)
plt.xlabel('Position')
plt.ylabel('Height')
plt.legend()
plt.show()
```

Pemain Tertinggi adalah GK sedangkan pemain terpendek adalah CMF

Grafik BoxpLot untuk melihat persebaran data dan mencari outliers

```python
sns.boxplot(x = dataset_pes["rating_stars"],y = dataset_pes["age"])
plt.title('Grafik Box Plot',fontsize = 25 , fontweight = 'bold')
plt.xlabel('Rating Stars')
plt.ylabel('Age')
plt.show()
```

Pemain dengan usia tertua dan termuda berada pada rating 1
