Informasi jumlah record, nama attribute, jumlah attribute, tipe objek, jumlah attribute, tipe angka, jumlah attribute yang memiliki nilai null ke dalam Dictionary

```python
jumlahRecord = len(dataset_pes)
namaAttribute = list(dataset_pes.columns)
jumlahAttributeObject = len(dataset_pes.select_dtypes(include = ["object"]).columns)
jumlahAttributeAngka = len(dataset_pes.select_dtypes(include = ["number"]).columns)
jumlahAttributeNull = len(dataset_pes.columns[dataset_pes.isna().sum() > 0])

dictionaryPes = {
    "Jumlah Record             :": jumlahRecord,
    "Nama Attribute            :": namaAttribute,
    "Jumlah Attribute Object   :": jumlahAttributeObject,
    "Jumlah Attribute Angka    :": jumlahAttributeAngka,
    "Jumlah Attribute Null     :": jumlahAttributeNull,
}

pd.Series(dictionaryPes)
```
